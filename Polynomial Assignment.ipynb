```python
# Step 1: Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score, mean_squared_error

# Step 2: Load the Dataset
# The dataset is small, with 10 rows and 3 columns: Position, Level, and Salary
data = pd.read_csv('G:/TOP-MENTOR/FILES/August 25th/Assignment/Polynomial Assignment/Position_Salaries.csv')

# Check the dataset structure
print(data.head())

# Step 3: Prepare Data for Regression
# Use 'Level' as the independent variable (X) and 'Salary' as the dependent variable (y)
X = data['Level'].values.reshape(-1, 1)
y = data['Salary'].values

# Step 4: Build and Fit a Linear Regression Model
linear_model = LinearRegression()
linear_model.fit(X, y)

# Predict salary using the Linear Regression model
y_pred_linear = linear_model.predict(X)

# Step 5: Visualize the Linear Regression Results
plt.scatter(X, y, color='red')  # Plot actual data points
plt.plot(X, y_pred_linear, color='blue')  # Plot the regression line
plt.title('Linear Regression: Level vs Salary')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()

# Step 6: Polynomial Regression
# Transform the data into polynomial features (degree=4, but can be adjusted)
poly = PolynomialFeatures(degree=4)
X_poly = poly.fit_transform(X)

# Fit the Polynomial Regression model
poly_model = LinearRegression()
poly_model.fit(X_poly, y)

# Predict salary using the Polynomial Regression model
y_pred_poly = poly_model.predict(X_poly)

# Step 7: Visualize the Polynomial Regression Results
plt.scatter(X, y, color='red')  # Actual data points
plt.plot(X, y_pred_poly, color='green')  # Polynomial regression curve
plt.title('Polynomial Regression: Level vs Salary')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()

# Step 8: Model Diagnostics - Evaluate Accuracy and Errors
# Linear Regression Performance
linear_r2 = r2_score(y, y_pred_linear)
linear_mse = mean_squared_error(y, y_pred_linear)

# Polynomial Regression Performance
poly_r2 = r2_score(y, y_pred_poly)
poly_mse = mean_squared_error(y, y_pred_poly)

# Print the diagnostics for both models
print(f'Linear Regression R: {linear_r2:.3f}, MSE: {linear_mse:.2f}')
print(f'Polynomial Regression R: {poly_r2:.3f}, MSE: {poly_mse:.2f}')
```
